```python
## 儲存參數程式
import pickle
import os

def save_variable(v,filename):
  f=open(filename,'wb')
  pickle.dump(v,f)
  f.close()
  return filename

def load_variavle(filename):
  f=open(filename,'rb')
  r=pickle.load(f)
  f.close()
  return r

# filename = save_variable(results,'results.txt')  ##儲存參數
# results = load_variavle('results.txt')  ##讀取參數

os.chdir('D:\\WEI\\TWStock')
```

```python
import os
data_date = '2021-11-17'  ## 手動輸出

## 讀取stock_result
patho = os.getcwd()
path = patho + '\stock_result\\' + data_date
os.chdir(path)
twse_stock_result = load_variavle('twse_stock_result.txt')
tpex_stock_result = load_variavle('tpex_stock_result.txt')
os.chdir(patho)

## 讀取check_result
path = patho + '\check_result\\' + data_date
os.chdir(path)
twse_check_result = load_variavle('twse_check_result.txt')
tpex_check_result = load_variavle('tpex_check_result.txt')
os.chdir(patho)
```

```python
## 篩選股票 厚任
def stock_choseH(codes,checks):
    stock_chosenH = []
    for i in range(len(codes)):
        code = codes[i]
        check = checks[i]
        
        check_n0 = check[0]
        check_n1 = check[1]
        check_n2 = check[2]
        check_n3 = check[3]
        check_n4 = check[4]

        chk_n0_volume = check_n0[1]
        chk_n0_vma5 = check_n0[6]
        chk_n0_vma10 = check_n0[7]
        
        chk_n0_ma5 = check_n0[2]
        chk_n1_ma5 = check_n1[2]
        chk_n2_ma5 = check_n2[2]
        chk_n3_ma5 = check_n3[2]
        chk_n4_ma5 = check_n4[2]
        
        chk_n0_ma10 = check_n0[3]
        chk_n1_ma10 = check_n1[3]
        chk_n2_ma10 = check_n2[3]
        chk_n3_ma10 = check_n3[3]
        chk_n4_ma10 = check_n4[3]
        
        chk_n0_ma20 = check_n0[4]
        chk_n1_ma20 = check_n1[4]
        chk_n2_ma20 = check_n2[4]
        chk_n3_ma20 = check_n3[4]
        chk_n4_ma20 = check_n4[4]
        
        chk_n0_ma60 = check_n0[5]
        chk_n1_ma60 = check_n1[5]
        chk_n2_ma60 = check_n2[5]
        chk_n3_ma60 = check_n3[5]
        chk_n4_ma60 = check_n4[5]

        chosen = []
        cond0 = min(chk_n0_ma5,chk_n0_ma10,chk_n0_ma20,chk_n0_ma60)
        cond1 = min(chk_n1_ma5,chk_n1_ma10,chk_n1_ma20,chk_n1_ma60)
        cond2 = min(chk_n2_ma5,chk_n2_ma10,chk_n2_ma20,chk_n2_ma60)
        cond3 = min(chk_n3_ma5,chk_n3_ma10,chk_n3_ma20,chk_n3_ma60)
        cond4 = min(chk_n4_ma5,chk_n4_ma10,chk_n4_ma20,chk_n4_ma60)
        cond_all = min(cond0,cond1,cond2,cond3,cond4)
        if cond_all > 0 and chk_n0_vma5 < 250 and chk_n0_volume > 1000 and chk_n0_ma5 < 7:
            chosen = [code,check_n0]
            stock_chosenH.append(chosen)
                
    return stock_chosenH
```

```python
## 厚任
## 篩選上市股票
codes = list(twse_check_result.keys())
checks = list(twse_check_result.values())
stock_chosenH_twse = stock_choseH(codes,checks)

## 篩選上櫃股票
codes = list(tpex_check_result.keys())
checks = list(tpex_check_result.values())
stock_chosenH_tpex = stock_choseH(codes,checks)


## 儲存上市上櫃chosenH成果到電腦
import os
patho = os.getcwd() 
os.chdir(os.getcwd() + '\stock_chosen')

pathw = os.getcwd() + '\\' + str(data_date)
if not os.path.isdir(pathw):
    os.mkdir(pathw)

os.chdir(pathw)
save_variable(stock_chosenH_twse,'stock_chosenH_twse.txt')
save_variable(stock_chosenH_tpex,'stock_chosenH_tpex.txt')
os.chdir(patho)
```
