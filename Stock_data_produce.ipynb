```python
import os
import numpy as np
import json


patho = os.getcwd()
path = patho + '\stock_chosen\\'
date_list = os.listdir(path)
os.chdir(patho)

for i in date_list:
    data_date = i
    twse_chosen_result, tpex_chosen_result,\
    twse_chosen2_result, tpex_chosen2_result,\
    twse_chosen3_result, tpex_chosen3_result = readChosen(data_date)
    
    chosen_data1,chosen_data2,chosen_data3 = chosen_data_covert(twse_chosen_result, tpex_chosen_result,\
                                                                twse_chosen2_result, tpex_chosen2_result,\
                                                                twse_chosen3_result, tpex_chosen3_result)
    
    saveChsenDic(data_date,chosen_data1,chosen_data2,chosen_data3)
```

```python
## 讀取chosen
def readChosen(data_date):
    patho = os.getcwd()
    path = patho + '\stock_chosen\\' + data_date
    os.chdir(path)
    twse_chosen_result = np.load('stock_chosen_twse.npy', allow_pickle='TRUE')
    tpex_chosen_result = np.load('stock_chosen_tpex.npy', allow_pickle='TRUE')
    twse_chosen2_result = np.load('stock_chosen2_twse.npy', allow_pickle='TRUE')
    tpex_chosen2_result = np.load('stock_chosen2_tpex.npy', allow_pickle='TRUE')
    twse_chosen3_result = np.load('stock_chosen3_twse.npy', allow_pickle='TRUE')
    tpex_chosen3_result = np.load('stock_chosen3_tpex.npy', allow_pickle='TRUE')
    os.chdir(patho)
    
    return twse_chosen_result, tpex_chosen_result, twse_chosen2_result, tpex_chosen2_result, twse_chosen3_result, tpex_chosen3_result 
```

```python
def chosen_data_covert(twse_chosen_result, tpex_chosen_result,\
                       twse_chosen2_result, tpex_chosen2_result,\
                       twse_chosen3_result, tpex_chosen3_result):
    
    chosen_data1 = []
    for i in range(len(twse_chosen_result)):
        chosen_dic = {}
        chosen_dic["stockId"] = twse_chosen_result[i][0]
        chosen_dic["stockPrice"] = twse_chosen_result[i][1][0]
        chosen_dic["stock10ma"] = twse_chosen_result[i][1][3]
        chosen_dic["stock20ma"] = twse_chosen_result[i][1][4]
        chosen_dic["stock60ma"] = twse_chosen_result[i][1][5]
        chosen_data1.append(chosen_dic)
    for i in range(len(tpex_chosen_result)):
        chosen_dic = {}
        chosen_dic["stockId"] = tpex_chosen_result[i][0]
        chosen_dic["stockPrice"] = tpex_chosen_result[i][1][0]
        chosen_dic["stock10ma"] = tpex_chosen_result[i][1][3]
        chosen_dic["stock20ma"] = tpex_chosen_result[i][1][4]
        chosen_dic["stock60ma"] = tpex_chosen_result[i][1][5]
        chosen_data1.append(chosen_dic)

    chosen_data2 = []
    for i in range(len(twse_chosen2_result)):
        chosen_dic = {}
        chosen_dic["stockId"] = twse_chosen2_result[i][0]
        chosen_dic["stockPrice"] = twse_chosen2_result[i][1][0]
        chosen_dic["stock10ma"] = twse_chosen2_result[i][1][3]
        chosen_dic["stock20ma"] = twse_chosen2_result[i][1][4]
        chosen_dic["stock60ma"] = twse_chosen2_result[i][1][5]
        chosen_data2.append(chosen_dic)
    for i in range(len(tpex_chosen2_result)):
        chosen_dic = {}
        chosen_dic["stockId"] = tpex_chosen2_result[i][0]
        chosen_dic["stockPrice"] = tpex_chosen2_result[i][1][0]
        chosen_dic["stock10ma"] = tpex_chosen2_result[i][1][3]
        chosen_dic["stock20ma"] = tpex_chosen2_result[i][1][4]
        chosen_dic["stock60ma"] = tpex_chosen2_result[i][1][5]
        chosen_data2.append(chosen_dic)

    chosen_data3 = []
    for i in range(len(twse_chosen3_result)):
        chosen_dic = {}
        chosen_dic["stockId"] = twse_chosen3_result[i][0]
        chosen_dic["stockPrice"] = twse_chosen3_result[i][1][0]
        chosen_dic["stock10ma"] = twse_chosen3_result[i][1][3]
        chosen_dic["stock20ma"] = twse_chosen3_result[i][1][4]
        chosen_dic["stock60ma"] = twse_chosen3_result[i][1][5]
        chosen_data3.append(chosen_dic)
    for i in range(len(tpex_chosen3_result)):
        chosen_dic = {}
        chosen_dic["stockId"] = tpex_chosen3_result[i][0]
        chosen_dic["stockPrice"] = tpex_chosen3_result[i][1][0]
        chosen_dic["stock10ma"] = tpex_chosen3_result[i][1][3]
        chosen_dic["stock20ma"] = tpex_chosen3_result[i][1][4]
        chosen_dic["stock60ma"] = tpex_chosen3_result[i][1][5]
        chosen_data3.append(chosen_dic)
        
    return chosen_data1, chosen_data2, chosen_data3
```

```python
## 儲存上市上櫃chosen_dic、chosen_dic2、chosen_dic3成果到電腦

def saveChsenDic(data_date,chosen_data1,chosen_data2,chosen_data3):
    patho = os.getcwd() 
    os.chdir(os.getcwd() + '\stock_json')

    pathw = os.getcwd() + '\\' + str(data_date)
    if not os.path.isdir(pathw):
        os.mkdir(pathw)

    os.chdir(pathw)
    with open("chosen_data1.json", "w") as f:
        json.dump(chosen_data1, f)
    with open("chosen_data2.json", "w") as f:
        json.dump(chosen_data2, f)
    with open("chosen_data3.json", "w") as f:
        json.dump(chosen_data3, f)
    os.chdir(patho)
```
